```python
import pandas as pd
import matplotlib.pyplot as plt

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("Jumlah variabel numerik:", len(numeric_cols))
print(numeric_cols)


for col in numeric_cols:
    plt.figure(figsize=(7,4))
    plt.hist(df[col].dropna(), bins=20)
    plt.title(f"Distribusi {col}")
    plt.xlabel(col)
    plt.ylabel("Frekuensi")
    plt.tight_layout()
    plt.show()
```

```python
latest_year = df['tahun'].max()
df_latest = df[df['tahun'] == latest_year]

print("Tahun terbaru dalam data:", latest_year)

# --- Fungsi plot top-bottom 5 ---
def plot_top_bottom_5(df, col, title, ylabel):
    top5 = df.nlargest(5, col)
    bottom5 = df.nsmallest(5, col)

    # Gabungkan & tandai
    tb = pd.concat([top5.assign(rank='Top 5'),
                    bottom5.assign(rank='Bottom 5')])

    plt.figure(figsize=(10,6))
    plt.barh(tb["kabupaten/kota"], tb[col])
    plt.title(title)
    plt.xlabel(ylabel)
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()


plot_top_bottom_5(
    df_latest,
    col="stunting_(%)",
    title=f"Top 5 & Bottom 5 Stunting – Tahun {latest_year}",
    ylabel="Persentase Stunting"
)

plot_top_bottom_5(
    df_latest,
    col="ipm",
    title=f"Top 5 & Bottom 5 IPM – Tahun {latest_year}",
    ylabel="Nilai IPM"
)

plot_top_bottom_5(
    df_latest,
    col="kemiskinan_(%)",
    title=f"Top 5 & Bottom 5 Kemiskinan – Tahun {latest_year}",
    ylabel="Persentase Kemiskinan"
)
```

```python
stunting_col = "stunting_(%)"
tahun_col = "tahun"

# Hitung rata-rata stunting per tahun seluruh kab/kota
trend = df.groupby(tahun_col)[stunting_col].mean().reset_index()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(trend[tahun_col], trend[stunting_col], marker='o', linewidth=2)

plt.title("Trend Stunting di Jawa Timur per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Rata-rata Stunting (%)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Cetak tabel ringkas
print(trend)
```

```python
trend_ipm = df.groupby("tahun")["ipm"].mean().reset_index()

plt.figure(figsize=(10,6))
plt.plot(trend_ipm["tahun"], trend_ipm["ipm"], marker='o', linewidth=2)
plt.title("Trend IPM Jawa Timur per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Rata-rata IPM")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("Trend IPM per Tahun:")
print(trend_ipm)


trend_ipm["delta"] = trend_ipm["ipm"].diff()
print("\nPerubahan IPM tiap tahun (delta):")
print(trend_ipm)


```

```python

trend_both = df.groupby("tahun")[["ipm", "stunting_(%)"]].mean().reset_index()

# Plot hubungan IPM vs Stunting (scatter)
plt.figure(figsize=(10,6))
plt.scatter(trend_both["ipm"], trend_both["stunting_(%)"])
plt.title("Hubungan IPM dan Stunting (Rata-rata per Tahun)")
plt.xlabel("IPM (Rata-rata Jatim)")
plt.ylabel("Stunting (%)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Hitung korelasi
corr = trend_both["ipm"].corr(trend_both["stunting_(%)"])
print(f"\nKorelasi IPM vs Stunting: {corr:.4f}")
```

```python
sns.set(style="whitegrid")

# =====================================================
# 1. HEATMAP STUNTING (Kabupaten × Tahun)
# =====================================================
pivot_stunting = df.pivot_table(
    index="kabupaten/kota",
    columns="tahun",
    values="stunting_(%)",
    aggfunc="mean"
)

plt.figure(figsize=(14, 12))
sns.heatmap(
    pivot_stunting,
    cmap="Reds",
    linewidths=0.3,
    linecolor="gray",
    annot=False
)
plt.title("Heatmap Stunting (%) per Kabupaten/Kota per Tahun", fontsize=14)
plt.xlabel("Tahun")
plt.ylabel("Kabupaten/Kota")
plt.tight_layout()
plt.show()

# =====================================================
# 2. HEATMAP IPM (Kabupaten × Tahun)
# =====================================================
pivot_ipm = df.pivot_table(
    index="kabupaten/kota",
    columns="tahun",
    values="ipm",
    aggfunc="mean"
)

plt.figure(figsize=(14, 12))
sns.heatmap(
    pivot_ipm,
    cmap="Greens",
    linewidths=0.3,
    linecolor="gray",
    annot=False
)
plt.title("Heatmap Indeks Pembangunan Manusia (IPM) per Kabupaten/Kota per Tahun", fontsize=14)
plt.xlabel("Tahun")
plt.ylabel("Kabupaten/Kota")
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

# Daftar pasangan variabel untuk scatter plot
pairs = [
    ("ipm", "stunting_(%)", "Stunting vs IPM"),
    ("kemiskinan_(%)", "stunting_(%)", "Stunting vs Kemiskinan"),
    ("tanpa_air_bersih_(%)", "stunting_(%)", "Stunting vs Tanpa Air Bersih"),
    ("pengeluaran_pangan_(%)", "stunting_(%)", "Stunting vs Pengeluaran Pangan"),
    ("rata_rata_lama_sekolah", "ipm", "IPM vs Rata-Rata Lama Sekolah"),
    ("kemiskinan_(%)", "ipm", "IPM vs Kemiskinan"),
]

# Loop untuk membuat scatter plot otomatis
for x, y, title in pairs:
    plt.figure(figsize=(8,6))
    sns.scatterplot(
        data=df,
        x=x,
        y=y,
        hue="tahun",
        palette="viridis",
        alpha=0.8
    )
    plt.title(title, fontsize=14)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend(title="Tahun")
    plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Hitung trend
trend_stunting = df.groupby("tahun")["stunting_(%)"].mean().reset_index()
trend_ipm = df.groupby("tahun")["ipm"].mean().reset_index()

# Gabungkan
trend = trend_stunting.merge(trend_ipm, on="tahun")
trend = trend.sort_values("tahun").reset_index(drop=True)


# Normalisasi 0-1
scaler = MinMaxScaler()
trend[["stunting_norm", "ipm_norm"]] = scaler.fit_transform(
    trend[["stunting_(%)", "ipm"]]
)

# Plot
plt.figure(figsize=(12,6))

plt.plot(trend["tahun"], trend["stunting_norm"], marker="o", linewidth=3, label="Stunting (Normalized)")
plt.fill_between(trend["tahun"], trend["stunting_norm"], alpha=0.15)

plt.plot(trend["tahun"], trend["ipm_norm"], marker="o", linewidth=3, label="IPM (Normalized)")
plt.fill_between(trend["tahun"], trend["ipm_norm"], alpha=0.15)

plt.title("Trend Stunting dan IPM (Normalisasi 0–1)")
plt.xlabel("Tahun")
plt.ylabel("Nilai Normalisasi (0–1)")
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```
