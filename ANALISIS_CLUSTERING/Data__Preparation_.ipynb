```python
import pandas as pd
stunting = pd.read_csv("/content/drive/MyDrive/StuntingdanIPM/fvsa_stanting.csv")
ipm = pd.read_csv("/content/drive/MyDrive/StuntingdanIPM/indikator_ipm.csv")
ipm2 = pd.read_csv("/content/drive/MyDrive/StuntingdanIPM/DATA_IPM_STUNTING_2020_2024 - ipm_long_form (1).csv")
```

```python
print("=== STUNTING DATA ===")
print(stunting.head())
print(stunting.info())

print("\n=== IPM DATA ===")
print(ipm.head())
print(ipm.info())
```

```python
stunting.columns = stunting.columns.str.strip().str.lower().str.replace(' ', '_')
ipm.columns = ipm.columns.str.strip().str.lower().str.replace(' ', '_')
ipm2.columns = ipm2.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
def fix_decimal(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].astype(str).str.replace(',', '.', regex=False)
            # ubah ke float bila mungkin
            df[col] = pd.to_numeric(df[col], errors='ignore')
    return df

stunting = fix_decimal(stunting)
ipm = fix_decimal(ipm)
ipm2 = fix_decimal(ipm2)
```

```python
ipm = ipm.rename(columns={
    'umur_harapan_hidup_saat_lahir': 'angka_harapan_hidup',
    'harapan_lama_sekolah': 'harapan_lama_sekolah',
    'rata-rata_lama_sekolah': 'rata_rata_lama_sekolah',
    'produk_domestik_regional_bruto_per_kapita_atas_dasar_harga_berlaku_menurut_kabupaten/kota_di_provinsi_jawa_timur_(ribu_rupiah)': 'pdrb_perkapita_ribu'
})
ipm2 = ipm2.rename(columns={'indeks_membangunan_manusia_(hasil_long_form_sp2020)_menurut_kab./kota': 'ipm'})
```

```python
ipm2
```

```python
stunting['kabupaten/kota'] = stunting['kabupaten/kota'].str.lower().str.strip()
ipm['kabupaten/kota'] = ipm['kabupaten/kota'].str.lower().str.strip()
ipm2['kabupaten/kota'] = ipm2['kabupaten/kota'].str.lower().str.strip()
```

```python
# Merge pertama: Stunting + IPM 2020-2024
merged1 = pd.merge(
    stunting,
    ipm,
    on=['kabupaten/kota', 'tahun'],
    how='left'
)

# Merge kedua: hasil merge pertama + IPM SP2020 Longform
merged2 = pd.merge(
    merged1,
    ipm2,
    on=['kabupaten/kota', 'tahun'],
    how='left',
    suffixes=('_old', '_new')
)

merged2.head()
```

```python
merged2.info()
```

```python
merged2 = merged2.drop(columns=["kabupaten/kota_code_y", "kabupaten/kota_code","sumber"])

```

```python
lokasi = pd.read_csv("/content/drive/MyDrive/StuntingdanIPM/Koordinat_Jawa_Timur.xlsx - Sheet1.csv")
lokasi
```

```python
import pandas as pd

# Jika ingin merge pakai kode
# Pastikan tipe datanya sama, misal int
merged2['kabupaten/kota_code_x'] = merged2['kabupaten/kota_code_x'].astype(int)
lokasi['Kode'] = lokasi['Kode'].astype(int)

# Merge berdasarkan kode kabupaten/kota
merged_final = pd.merge(
    merged2,
    lokasi[['Kode', 'Latitude', 'Longitude']],
    left_on='kabupaten/kota_code_x',
    right_on='Kode',
    how='left'
)

# Hapus kolom Kode dari lokasi agar tidak duplikat
merged_final = merged_final.drop(columns=['Kode'])

# Cek hasil
merged_final.info()
merged_final.head()
```
